```python
#给定一个非负整数 num，反复将各个位上的数字相加，直到结果为一位数。返回这个结果
#结果一定是除余9,所得结果，因为当十位上的数以个位计算时，相当于每个10里抽一个1
#比如83-->8+3    80变成8，相当于8个10，每个10减去9，剩下一个1得到8
#百位千位等等同理，百位是每个100抽一个1，千位是每个1000抽一个1
if not num: return 0#当num为0时，是False，则not False返回0
return 9 if num%9==0 else num%9#当num能被除尽时，最后剩余的个位一定是9，当不能被除尽时，余数就是最后所得
#或者
return (num-1)%9+1 if num else 0#当num为0时，是False，则返回0，当不是0，则是True,返回(num-1)%9+1
#num-1是为了让数余下，再加回来可以还原余数
```

```python
#丑数 就是只包含质因数 2、3 和 5 的 正 整数。
#给你一个整数 n ，请你判断 n 是否为 丑数 。如果是，返回 true ；否则，返回 false 。
#显然，丑数就是只能被2，3，5除，并且除完剩下的一定是个1，即丑数n=2**a*3**b*5**c    a,b,c为自然数，当且仅当a=b=c=0时，丑数为1成立
#思路就是，可以让丑数按顺序除2，3，5，直到把其中一个数除尽，再除下一个数
if n<1: return False
while n%2==0:
    n=n//2#也能写成n//=2
while n%3==0:
    n=n//3
while n%5==0:
    n=n//5
return n==1#如果n==1则输出True否则输出False
```
