```python
#[[fill]align][sing][#][0][width][grouping_option][.precision][type]
#fill即索引
#align,对齐方式,用0补齐时，0会自动补在数字与正负号之间，记住负号也算一个字符
#"^"中对齐
#“<,>”尖尖朝哪向哪对齐

#sign
#“+”表示总是显示符号，即在正数前面添加正号（+），在负数前面添加负号（-）
#“-”表示仅显示负号，即只有负数在前面添加符号（-），默认行为，也可以不写
#“ ”空格，表示在数字前加东西，即在正数前面添加一个空格，在负数在前面添加负号（-）
#","和“_”都是千分符,","指用“，”分，“_”指用“_”分位数不够不显示千分符

#[.precision]精度选项：只需要填入一个十进制整数，对不同类型参数效果不一样,格式为{:.数字type}
#对于[typr]设置为"f"或"F"的浮点数来说，是限定小数点后显示多少个数位
#对于[type]设置为"g"或"G"的浮点数来说，是限定小数点前后一共显示多少数位
#对于非数字类型来说，限定的是最大字段的大小
#对于整数类型来说，则不允许使用这个精度选项，会报错

#type类型选项1
#“b”将参数以二进制的形式输出
#“c”将参数以Unicode的形式输出
#“d”将参数以十进制的形式输出
#“o”将参数以八进制的形式输出
#“x或X”将参数以十六进制的形式输出
#“h”跟“d”类似，不同之处在于它会使用当前语言环境设置的分隔符插入到恰当位置
#“None”跟“d”一样
#这八个适用于整数
#可以在这之前加一个“#”，可以显示一个前缀，表明他是几进制，比如十六进制就会有前缀0X

#type类型选项2
#“e或E”将参数以科学计数法的形式输出，（以字母“e或E”来标志指数，默认精度为6）
#“f或F”将参数以定点表示法的形式输出，（“不是数”用nan或NAN标志，无穷用“inf或INF”标志，默认精度为6）
#“g或G”通用格式，小数以“f或F”形式输出，大数以“e或E”的形式输出
#“n”跟“g”类似，不同之处在于它会使用当前语言环境设置的分隔符插入到恰当位置
#“%”以百分比形式输出（将数字乘以100并显示为定点表示法（“f”）的形式，后面带一个“%”）
#“None”跟“g”类似，不同之处在于使用定点表示法时，小数点后将至少显示一位；默认精度与给定值所需精度一致
#适用于浮点数或复数，即使输入整数，也会自动转换为等值浮点数

#支持用关键字来索引这些值，如
print("{:.{i}f}".format(3.1415,i=2))
print("{:{i}{m}{n}.{j}{k}}".format(3.1415,i="+",m="^",n=10,j=2,k="f"))
```

```python
#字典(dict)——直接体现编程内部的映射关系,格式{ : },键值对，“：”左侧叫键，右侧叫值,键不能重复，键不可修改
#字典可迭代，只是默认是迭代键，可通过加后缀实现迭代值或键值对
#创建字典,常见有6种方法
a={"吕布":"吕小布","关羽":"关二爷","刘备":"草鞋"}#直接定义
b=dict(吕布="吕小布",关羽="关二爷",刘备="草鞋")#创建字典，要求键不能带引号
c=dict([("吕布","吕小布"),("关羽","关二爷"),("刘备","草鞋")])#这样也能创建出键，由列表和元组合成
d=dict({'吕布': '吕小布', '关羽': '关二爷', '刘备': '草鞋'})#这样也行
e=dict({'吕布': '吕小布', '关羽': '关二爷',},刘备="草鞋")#混合用法
f=dict(zip(["吕布","关羽","刘备"],["吕小布","关二爷","草鞋"]))#用序列中的zip调用俩列表也行
print(a,b,c,d,e,f)
print(a==b==c==d==e==f)
#以上六种
```

```python
#函数：打包代码
#创建和调用函数,用def
#调用函数时，不论有没有参数，末尾都得带括号，不带括号是引用函数，打印出的是函数的类似地址的东西，带括号才是调用函数，输出函数运行结果
#函数的参数：支持多参数，但是要对应
#函数的参数还分为形式参数和实际参数，即形参和实参
#形参：就是创建函数时用的变量的名字，比如创建函数def i(name):  时，name就是形参
#实参：就是调用函数时，要代替形参的具体值，比如调用函数时i("Python")   的“Python”代替了函数中的“name”，这个Python就是实参
#占位置的参数叫做形参，实际调用的参数叫做实参
#函数的返回值，用return语句实现,返回时，实参是按照形参的顺序返回的
#函数遇到“return”时立刻直接就返回了，不会理后边的语句，也不会受后边的语句的影响
#如果函数没用“return”则会在执行完函数体之后自行偷偷的返回一个“None”
```

```python
#函数的参数：
#1.位置参数：实参的输出顺序由形参的顺序决定，这种位置固定的参数被称为位置参数
#2.关键字参数：这种调用函数时直接用形参="实参"的方法输出的实参，叫做关键字参数
#当关键字参数和位置参数混合使用时，位置参数必须在关键字参数之前，而且地保证用位置参数的实参位置必须与对应形参相同，否则报错，位置参数对位置匹配要求极高
#3.默认参数：允许在设置形参时就用 形参="默认参数" 的方法设置，默认参数，当调用函数时不给有默认参数的形参输入实参时，就会输出默认参数，用默认参数时，要放在最后，不然会报错
#return返回函数时，可以返回表达式等等，注意不能返回"语句",如不能返回赋值语句，循环语句，条件语句等等
```

```python
#收集参数:支持参数可多可少的函数，比如print(),括号内几个参数都可以被打印
#创建函数时，可以在形参前加上"*",比如def i(*r):,此时形参"r"就会变成收集参数，支持实参的可多可少吗，吸收所有的位置参数
#简单来说就是用户像输入几个实参都可以，其实，收集参数的实质就是元组，通过元组的打包与解包实现实现存储打印多参数
#还记得之前刚讲打包和解包的时候，举例子说是一般情况下，要被打包的字符串元素数量必须与参数数量相同，除非用"*",且在最后一个。这样剩下的就会全都储存在最后那个带"*"的参数中，其实最后那个参数就是收集参数
#如果收集参数后还有要设实参的形参，那么，后边的参数必须用关键字参数，不然后边的实参都会被收集参数吸收，要不然就把收集参数放最后
#收集参数不只可以被设为元组，一个"*"是元组，而用两个连续的"*"就是字典
#因为字典内元素都是键值对，所以实参必须是关键字参数
#但是不是普通的关键字参数，是关键字参数收集，就是创建函数时，不是普通的形参，而是字典型收集参数，然后调用函数时，实参也不是普通的实参，是由"未定义形参=实参"，的形式的实参，未定义形参收入字典中成为键，实参成为值
#普通形参，元组型收集参数和字典型收集参数可以混合使用
#format()函数是自带两种收集参数的天选之子
#"*"和"**"在创建函数的形参中是打包功能，而在调用函数的实参中是解包功能
```

```python
#作用域：指一个变量可以被访问的范围，一个变量的作用域由它被赋值的位置决定
#局部作用域：如果一个变量定义的位置在一个函数中，那它的作用域就仅限该函数中
#全局作用域（一般称为全局变量）：在外部定义的变量，在哪都能调用打印
#一般情况下，函数可以调用全局变量，但是无法改变全局变量，强行改变只会创建一个新的同名变量
#当内外部同时定义同名变量时，在函数内部，内部变量会覆盖外部变量，而在函数外部，则会输出全局变量，而且，虽然俩变量同名，但是不是同一个变量，他们的id值不一样
#当在外部设了一个全局变量时，当打算在函数内部修改这个变量时，会自动创建这个同名变量，独立于原变量的变量
```

```python
#global语句：在函数内外部设同名变量时，该语句可以实现防止在函数内部增加这个同名变量，也就是说，允许在函数内部改变这个全局变量的值，而不新建同名变量
#用法；直接在函数内部写一行'global 变量'就行
```

```python
#函数的嵌套：函数嵌套时，内部函数无法被直接调用,想调用内部函数，就必须在创建函数时，在外部函数里调用内部函数
#同样，一般情况下内部函数只能调用外部函数的变量，而无法改变这个变量，强行改变同样只会创建一个新的同名变量
#nonlocal语句:能使内部函数修改外部函数变量，类似于gobal语句，只是使用对象不一样，用法也和gobal语句类似
#对于嵌套函数来说，外层函数作用域被某种形式保存下来了，尽管外层函数已经调用完了，外层函数作用域变量也不会消失
```

```python
#LEGB规则：本地优先原则
#1.就是当同时存在全局变量和局域变量时，局域变量会覆盖全局变量
#2.当同时存在外局域变量和局域变量时，局域变量会覆盖外局域变量
#3.当内置函数名字（如str）等被占用时，如设str为一个变量或自定义函数时，那str将失去它转化字符串的能力，所以自定义变量时，尽量不要用内置函数名为变量
#1和2中的覆盖，只在局部覆盖，谁的主场谁是老大，除非用global语句和nonlocal语句
```

```python
#装饰器：在不修改原代码的情况下，增加功能，装饰器通常是以闭包的形式编写的
#支持把多个装饰器装在一个函数上，也就是一个函数支持同时加入多个装饰器
import time
def time_master(func):
    def call_func():
        print("开始运行程序...")
        start=time.time()
        func()
        stop=time.time()
        print("结束程序运行...")
        print(f"一共用了{stop-start:.2f}秒。")
    return call_func
def myfunc():
    time.sleep(2)
    print("Hello...")
myfunc=time_master(myfunc)
myfunc()#装饰器基本原理
```

```python
import time
def time_master(func):
    def call_func():
        print("开始运行程序...")
        start=time.time()
        func()
        stop=time.time()
        print("结束程序运行...")
        print(f"一共用了{stop-start:.2f}秒。")
    return call_func
@time_master
def myfunc():
    time.sleep(2)
    print("Hello...")
myfunc()#@time_master是装饰器的一个语法糖，这里的myfunc()不是调用函数Myfunc的意思，而是把myfunc作为参数，把并把call_func函数赋给它
#也就是说，这里的myfunc()其实是调用的call_func函数
```

```python
#装饰器会使原函数变化，可以再外层函数的内部创建内层函数之前加上"@functools.wraps(外层函数形参)"就可以把函数的东西复制给装饰器内部函数
import functools
def log(func):
    @functools.wraps(func)
    def inner(*args,**kwargs):
        result=func(*args,**kwargs)#保存实参
        print(f'调用{func.__name__}{args}{kwargs}->{result}')
        return result
    return inner#可以提高泛用性
@log
def add(a,b):
    return a+b
@log
def greet(name,greeting="Hello"):
    return f"{greeting},{name}!"
add(1,2)#调用add(1,2)实际是inner(1,2),所以装饰器里inner也要有形参，为了保险，要设为收集参数
greet('tom','Hi')
```

```python
#@div1
#@div2
#调用多个装饰器的时候，顺序是从下往上，如先进行div2,再进行div1
#调用多个装饰器的时候，如果内部装饰其需要参数，则形参必须是收集参数
def add(func):
    def inner():
        x=func()
        return x+1
    return inner
def cube(func):
    def inner():
        x=func()
        return x*x*x
    return inner
def square(func):
    def inner():
        x=func()
        return x*x
    return inner
@add
@cube
@square
def test():
    return 2
print(test())
```

```python
#如果想给装饰器参数，只需要在最外层嵌套一个函数闭包就行了
import time
def logger(msg):
    def time_master(func):
        def call_func():
            print("开始运行程序...")
            start=time.time()
            func()
            stop=time.time()
            print("结束程序运行...")
            print(f"{msg}一共用了{stop-start:.2f}秒。")
        return call_func
    return time_master
@logger(msg="A")
def funA():
    time.sleep(1)
    print("正在调用funA...")
@logger(msg="B")
def funB():
    time.sleep(1)
    print("正在调用funB...")
funA()
funB()
```

```python
#递归：就是函数调用自身的过程,追忆区分形参实参，然后注意结束条件，就能掌握递归
#递归效率较慢，但是帅气，迭代效率更快
#注意使用递归的时候，一定要设置结束条件，否则会进入死循环
def funA(i):
    if i==1:
        return 1
    return i*funA(i-1)
funA(5)#递归求阶乘
```

```python
#汉诺塔:三根柱子,x,y,z柱子x上有i个罗盘(从下到上越来越小),要把罗盘全移动到z上,一次只能动一个,大的必须在小的下面
def hanoi(n,x,y,z):
    if n==1:
        print(x,"-->",z)
    else:
        hanoi(n-1,x,z,y)
        print(x,"-->",z)
        hanoi(n-1,y,x,z)
n=int(input("层数："))
hanoi(n,'A','B','C')
#一定要分清其中的形参与实参
#外部的实参不一定是内部函数的实参，递归时，内部进入函数时的实参是外部函数的形参对应的实参的实参
#比如上式，例如n是3，则进入else里，又进入函数，此时实参是被赋完值的n,x,y,z，对应出来就是2，x是A，z是C，y是B
#没带"def"就是调用，此时调用的是hanoi(2,A,C,B),这是实参，对应形参，形参变成，n=2,x=A,y=C,z=B,再开始进行函数
#往下进行，依旧进else里，由此进入函数，此时调用的是hanoi(1,A,B,C)，对应出来变成了n=1,x=A,y=B,z=C
```
