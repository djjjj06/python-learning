```python
import random
counts=7
num1=0
num2=100
answer=random.randint(0,100)
while counts>0:
    while True:
        temp=input("不妨猜猜小甲鱼现在心里想的是0到100之间哪个整数：")
        if temp.isdigit():
            guess=int(temp)
            if guess==answer:
                print("你是小甲鱼心中的蛔虫吗？")
                print("哼，猜中了也没有奖励！")
                counts=0
            else:
                if guess<answer:
                    print("小了~")
                    num1=guess+1
                else:
                    print("大了~")
                    num2=guess-1
                print("还剩{}次机会,当前范围是[{},{}]".format(counts-1,num1,num2))
            counts=counts-1
            break
        else:
            print("请输入数字!")
if counts==0:
    print("答案是{}".format(answer))
print("游戏结束，不玩了^_^")
```

```python
while True:
    m=input("请设置密码(8位):")
    if len(m)!=8:
        print("请按要求设置8位密码！")
        continue
    has_upper=False
    for i in m:
        if i.isupper():
            has_upper=True
            break
    if not has_upper:
        print("请包含至少一个大写字母！")
        continue
    has_lower=False
    for i in m:
        if i.islower():
            has_lower=True
            break
    if not has_lower:
        print("请至少包含一个小写字母！")
        continue
    has_digit=False
    for i in m:
        if i.isdigit():
            has_digit=True
            break
    if not has_digit:
         print("请至少包含一个数字！")
         continue
    print("密码设置成功！")
    break
counts=3
while counts>0:
    n=input("请输入密码：")
    if n==m:
        print("登录成功！")
        break
    counts-=1
    if counts>0:
        print("密码错误，还剩{}次机会".format(counts))
    else:
        print("账号已锁定，请稍后再试！")
```

```python
contacts=[]
while True:
    i=input("1.添加联系人\n2.删除联系人\n3.修改电话\n4.查找联系人\n5.显示所有联系人\n6.退出\n请选择您需要的服务：\n")
    if i=="1":
        name1=input("请输入联系人姓名：")
        count=input("请输入电话号码：")
        while True:
            if not count.isdigit():
                print("请输入有效手机号！")
                continue
            people=[name1,count]
            has_name=False
            for re_people in contacts:
                if people==re_people:
                    print("已有该人物信息！")
                    has_name=True
                    break
            if not has_name:
                contacts.append(people)
            break
    elif i=="2":
        name2=input("请输入要删除的联系人姓名：")
        counts2=-1
        has_pop=False
        for re_name in contacts:
            counts2+=1
            if name2 in re_name:
                has_pop=True
                answer=input("确定要删除吗？(yes or no)")
                if answer=="yes":
                    contacts.pop(counts2)
                    print("删除成功！")
                    break
                elif answer=="no":
                    break
                else:
                    print("操作有误，强制取消操作!")
                    break
        if not has_pop:
            print("无该人物信息！")
    elif i=="3":
        name3=input("请输入要修改号码的人物昵称：")
        counts3=-1
        has_gai=False
        for re_name in contacts:
            counts3+=1
            if name3 in re_name:
                has_gai=True
                while True:
                    new_count=input("请输入新的手机号：")
                    if new_count.isdigit():
                        answer=input("确定要修改吗？(yes or no)")
                        if answer=="yes":
                            contacts[counts3]=[name3,new_count]
                            print("修改成功！")
                            break
                        elif answer=="no":
                            break
                        else:
                            print("操作有误，强制取消操作!")
                            break
                    else:
                        print("请输入有效手机号！")
        if not has_gai:
            print("无该人物信息！")
    elif i=="4":
        name4=input("请输入要查找的人物姓名：")
        has_same=False
        counts4=-1
        for re_name in contacts:
            counts4+=1
            if name4 in re_name:
                has_same=True
                print(contacts[counts4])
                break
        if not has_same:
            print("无该人物信息！")
    elif i=="5":
        print(contacts)
    elif i=="6":
        print("已退出")
        break
    else:
        print("请重新输入您的需求！")
```

```python
import random
i=random.randint(0,999)
i=str(i).zfill(3)
counts=5
while True:
    count1=0
    count2=0
    answer=input("请输入三位数字密码：")
    if len(answer)!=3:
        print("格式错误！请重新输入")
        continue
    if answer.isdigit():
        if answer==i:
            print("密码正确！")
            break
        counts-=1
        if counts>0:
            for m in range(3):
                if i[m]==answer[m]:
                    count1+=1
            count_i=[0]*10
            count_answer=[0]*10
            for ch in i:
                count_i[int(ch)]+=1
            for ch in answer:
                count_answer[int(ch)]+=1
            for k in range(10):
                count2+=min(count_i[k],count_answer[k])
            print("密码错误，还剩{}次机会，其中有{}个位置和数字均正确，有{}个数字正确但位置不正确".format(counts,count1,count2-count1))
        else:
            print("错误次数过多，已锁定")
            break
    else:
        print("请按要求输入数字！")
```

```python
print("学生成绩管理系统")
students=[]
while True:
    answer=input("1.添加学生\n2.删除学生\n3.修改成绩\n4.查询成绩\n5.显示所有学生（按成绩从高到低排序）\n6.统计（平均分，最高分，最低分）\n7.退出\n请输入您的需求：")
    if answer=="1":
        student_name=input("请输入要添加学生的姓名：")
        if student_name.isalpha():
            student_name=student_name.capitalize()
        has_name=False
        for name,score in students:
            if student_name==name:
                print("已有该学生信息！")
                has_name=True
                break
        if not has_name:
            while True:
                student_score=input("请输入学生成绩：")
                if student_score.isdigit() and 0<=int(student_score)<=100:
                    students.append((student_name,int(student_score)))
                    print("添加成功！")
                    break
                elif student_score=="q":
                    print("已返回！")
                    break
                else:
                    print("输入有误，请重新输入或输入\"q\"返回！")
    elif answer=="2":
        student_name=input("请输入要删除的学生姓名：")
        if student_name.isalpha():
            student_name=student_name.capitalize()
        has_student=False
        count1=-1
        for name,score in students:
            count1+=1
            if student_name==name:
                has_student=True
                break
        if not has_student:
            print("无该学生信息！")
        else:
            answer1=input("请确定是否要删除（yes or no）：")
            if answer1=="yes":
                students.pop(count1)
                print("删除成功！")
            elif answer1=="no":
                print("已取消删除！")
            else:
                print("输入有误，已返回！")
    elif answer=="3":
        student_name=input("请输入要修改成绩的学生姓名：")
        if student_name.isalpha():
            student_name=student_name.capitalize()
        has_score=False
        count2=-1
        for name,score in students:
            count2+=1
            if student_name==name:
                has_score=True
                break
        if not has_score:
            print("未找到该学生信息！")
        else:
            while True:
                new_score=input("请输入修改后的成绩：")
                if new_score.isdigit() and 0<=int(new_score)<=100:
                    answer2=input("请确定是否要修改（yes or no）：")
                    if answer2=="yes":
                        students[count2]=(name,int(new_score))
                        print("修改成功！")
                        break
                    elif answer2=="no":
                        print("已取消修改！")
                        break
                    else:
                        print("输入有误，已返回！")
                        break
                elif new_score=="q":
                    print("已返回！")
                    break
                else:
                    print("输入有误，请重新输入或输入\"q\"返回！")
    elif answer=="4":
        student_name=input("请输入要查询学生的姓名：")
        same_student=False
        if student_name.isalpha():
            student_name=student_name.capitalize()
        for name,score in students:
            if student_name==name:
                same_student=True
                print(name,":",score)
                break
        if not same_student:
            print("未找到该学生信息！")
    elif answer=="5":
        if not students:
            print("当前未录入任何学生信息！")
        else:
            students1=sorted(students,key=lambda x:x[1],reverse=True)
            for name,score in students1:
                print(f"姓名：{name}，成绩：{score}")
    elif answer=="6":
        if not students:
            print("暂无学生！")
        else:
            total=0
            max_score=0
            min_score=100
            for name,score in students:
                total+=score
                if score>max_score:
                    max_score=score
                if score<min_score:
                    min_score=score
            avg=total/len(students)
            print(f"平均数为{avg:.2f}，最高分为{max_score}，最低分为{min_score}")
    elif answer=="7":
        print("欢迎下次再来！")
        break
```

```python
print("购物车系统")
cart={}
while True:
    answer=input("1.添加单个商品(add)\n2.删除商品(pop)\n3.修改价格(redact)\n4.查看所有商品(look)\n5.计算总价(total)\n6.退出(exit)\n请输入您的需求：")
    if answer=="add":
        name=input("请输入要加入购物车的商品:")
        if name in cart:
            same_name=input("您的购物车中已有该商品，是否再次购买(yes or no)：")
            if same_name=="yes":
                while True:
                    price=input("请输入商品价格：")
                    if price=="q":
                        print("已返回！")
                        break
                    elif not price.isnumeric():
                        print("请输入正确价格！或输入“q”键退出")
                        continue
                    price=float(price)
                    cart[name]["数量"]+=1
                    cart[name]["价格"]=price
                    print("添加成功！")
                    break
            elif same_name=="no":
                print("已取消操作！")
            else:
                print("输入有误，已强制返回！")
        else:
            while True:
                price=input("请输入商品价格：")
                if price=="q":
                    print("已返回！")
                    break
                elif not price.isnumeric():
                    print("请输入正确价格！或输入“q”键退出")
                    continue
                price=float(price)
                cart[name]={"价格":price,"数量":1}
                print("添加成功！")
                break
    elif answer=="pop":
        name=input("请输入要删除的商品:")
        sure_or_no=input("您确定要删除该商品吗(yes or no)?")
        if sure_or_no=="yes":
            if name in cart:
                cart.pop(name)
                print("删除成功！")
            else:
                print("您的购物车中没有该商品！")
        elif sure_or_no=="no":
            print("已取消操作！")
        else:
            print("输入有误，已强制返回！")
    elif answer=="redact":
        name=input("请输入要修改价格的商品：")
        if name not in cart:
            print("您的购物车中暂无该商品！")
            continue
        sure_or_no=input("您确定要修改该商品的价格吗(yes or no)?")
        if sure_or_no=="yes":
            while True:
                new_price=input("请输入修改后的价格：")
                if new_price=="q":
                    print("已返回！")
                    break
                elif not new_price.isnumeric():
                    print("请输入正确价格！或输入“q”键退出")
                    continue
                new_price=float(new_price)
                cart[name]["价格"]=new_price
                print("修改成功！")
                break
        elif sure_or_no=="no":
            print("已取消操作！")
        else:
            print("输入有误，已强制返回！")
    elif answer=="look":
        for goods in cart.items():
            print(goods)
    elif answer=="total":
        totals=sum([name["价格"]*name["数量"] for name in cart.values()])
        print(f"当前购物车内总价为{totals:.2f}")
    elif answer=="exit":
        print("欢迎下次再来！")
        break
    else:
        print("请按要求输入您的需求！")
```

```python
def word_statistics(text,ignore_case=True):
    text=text.lower()
    text=text.center(len(text)+2)
    word=[]
    words=[]
    count={}
    for letter in text:
        if letter.isalpha():
            word.append(letter)
        else:
            if word:
                word="".join(word)
                words.append(word)
                word=[]
    for same_word in words:
        if same_word in count:
            count[same_word]+=1
            continue
        count[same_word]=1
    return sorted(count.items(),key=lambda x:(-x[1],x[0]))
text=input("请输入要统计的英语单词：")
for i,j in word_statistics(text):
    print(f"{i}:{j}")
```

```python
#汉诺塔:三根柱子,x,y,z柱子x上有i个罗盘(从下到上越来越小),要把罗盘全移动到z上,一次只能动一个,大的必须在小的下面
def hanoi(n,x,y,z):
    if n==1:
        print(x,"-->",z)
    else:
        hanoi(n-1,x,z,y)
        print(x,"-->",z)
        hanoi(n-1,y,x,z)
n=int(input("层数："))
hanoi(n,'A','B','C')
#一定要分清其中的形参与实参
```

```python
import time
import functools
def time_cache(maxsize=128):
    def outer(func):
        i={}
        @functools.wraps(func)
        def inner(*args,**kwargs):
            n=args[0]
            if n not in i:
                start=time.time()
                result=func(*args,**kwargs)
                stop=time.time()
                if len(i)==maxsize:
                    del i[next(iter(i))]
                i[n]=result
                print(f"{func.__name__}{n}耗时：{stop-start:.4f}秒")
                return result
            print(f"缓存命中{func.__name__}{n}")
            return i[n]
        return inner
    return outer
@time_cache(maxsize=128)
def fib(n):
    if n==1 or n==2: return 1
    return fib(n-1)+fib(n-2)
n=int(input("输入要输出项数："))
fib(n)
```

```python
import time
import functools
def timer(func):
    @functools.wraps(func)
    def inner(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        stop=time.time()
        print(f"{func.__name__} 执行耗时：{stop-start}秒")
        return result
    return inner
@timer
def slow_sum(n):
    total=0
    for i in range(n):
        total+=i
    return total
slow_sum(1000000)
```

```python
import pickle
from datetime import datetime
records=[]
try:
    with open("个人收支记录.pkl","rb") as f:
        records=pickle.load(f)
        count=max(i['id'] for i in records) if records else 0
except FileNotFoundError:
    count=0
def fun1_date(date):
    try:
        datetime.strptime(date,"%Y-%m-%d")
        return True
    except ValueError:
        print(f"输入格式有误，请重新输入！")
        return False
print("===个人记账本系统===")
while True:
    try:
        answer=int(input("1.添加收支记录\n2.删除记录\n3.修改记录\n4.查询记录\n5.统计\n6.显示所有记录\n7.保存数据并退出\n请输入需求:"))
    except ValueError:
        print("请按要求输入需求！！！")
        continue
    if answer==1:
        while True:
            date=input("请输入日期(YYYY-MM-DD):")
            if fun1_date(date): break
        while True:
            category=input("请输入类别(类别可选:餐饮、购物、交通、娱乐、工资、其他):")
            if category not in ('餐饮','购物','交通','娱乐','工资','其他'):
                print("暂无该类别，请重新选择！")
        desc=input("请简述一下该收支出处或用途:")
        while True:
            try:
                amount=int(input("收支金额:"))
                break
            except ValueError:
                print("请输入正常金额！")
        count+=1
        records.append({'id':count,'date':date,'category':category,'desc':desc,'amount':amount})
    elif answer==2:
        if not records:
            print("当前无任何记录!")
            continue
        while True:
            try:
                need_remove=int(input("请输入要删除的记录编号:"))
                break
            except ValueError:
                print("请输入数字编号！")
        has_id=False
        for ind,i in enumerate(records):
            if i["id"]==need_remove:
                records.pop(ind)
                print("删除成功！")
                has_id=True
                break
        if not has_id:
            print("未找到该纪录！")
    elif answer==3:
        if not records:
            print("当前无任何记录!")
            continue
        while True:
            try:
                need_chance=int(input("请输入要修改的记录编号:"))
                break
            except ValueError:
                print("请输入数字编号！")
        has_id=False
        for ind,i in enumerate(records):
            if i["id"]==need_chance:
                while True:
                    need_chance_type=input("请选择要修改的项目(日期、类别、描述、金额):")
                    if need_chance_type in ('日期','类别','描述','金额'):
                        if need_chance_type=="日期":
                            while True:
                                date=input("请输入修改后的日期(YYYY-MM-DD):")
                                if fun1_date(date): break
                            i['date']=date
                            print(f"修改成功！")
                        elif need_chance_type=="类别":
                            while True:
                                category=input("请输入修改后的类别(类别可选:餐饮、购物、交通、娱乐、工资、其他):")
                                if category not in ('餐饮','购物','交通','娱乐','工资','其他'):
                                    print("暂无该类别，请重新选择！")
                                    continue
                            i['category']=category
                            print(f"修改成功！")
                        elif need_chance_type=="描述":
                            desc=input("请简述一下修改后该收支出处或用途:")
                            i['desc']=desc
                            print(f"修改成功！")
                        elif need_chance_type=="金额":
                            while True:
                                try:
                                    amount=int(input("收支金额:"))
                                    break
                                except ValueError:
                                    print("请输入正常金额！")
                                    continue
                            i['amount']=amount
                            print(f"修改成功！")
                        has_id=True
                        break
                    elif need_chance_type=="":
                        print("已取消修改！")
                        has_id=True
                        break
                    else:
                        print("无该项目！")
        if not has_id:
            print("未找到该纪录！")
    elif answer==4:
        if not records:
            print("当前无任何记录!")
            continue
        while True:
            try:
                need_chance=int(input("请输入要查询的记录编号:"))
                break
            except ValueError:
                print("请输入数字编号！")
        has_id=False
        for ind,i in enumerate(records):
            if i["id"]==need_chance:
                print(i)
                has_id=True
                break
        if not has_id:
            print("未找到该纪录！")
    elif answer==5:
        print("——————统计功能——————")
        if not records:
            print("当前无任何记录!")
            continue
        types=input("请输入统计方式(按月统计、按类别统计):")
        if types=="按月统计":
            while True:
                mouse_Z=[]
                mouse_F=[]
                mouse=input("请输入要统计的月份YYYY-MM:")
                if len(mouse)!=7 or mouse[4]!='-':
                    print("月份格式错误,请重新输入！")
                    continue
                for i in records:
                    if i['date'].startswith(mouse):
                        if i['amount']>=0: mouse_Z.append(i['amount'])
                        else: mouse_F.append(i['amount'])
                print(f"{mouse}总收入:{sum(mouse_Z)},总支出:{-sum(mouse_F)},结余:{sum(mouse_Z)-sum(mouse_F)}")
                break
        elif types=="按类别统计":
            while True:
                category_Z=[]
                category_F=[]
                typess=input("请输入类型(餐饮、购物、交通、娱乐、工资、其他):")
                if typess not in ('餐饮','购物','交通','娱乐','工资','其他'):
                    print("无该类别!")
                    continue
                for i in records:
                    if i['category']==typess:
                        if i['amount']>=0: category_Z.append(i['amount'])
                        else: category_F.append(i['amount'])
                print(f"{typess}总收入:{sum(category_Z)},总支出:{-sum(category_F)},结余:{sum(category_Z)-sum(category_F)}")
                break
        else:
            print("无该统计类型!")
    elif answer==6:
        if not records:
            print("当前无任何记录!")
            continue
        for i in records:
            print(i)
    elif answer==7:
        with open("个人收支记录.pkl","wb") as f:
            pickle.dump(records,f)
        print("欢迎下次再来！")
        break
```

```python
class Shape:
    def __init__(self,name):
        self.name=name
    def area(self):
        pass
class Square(Shape):
    def __init__(self,length):
        super().__init__("正方形")
        self.length=length
    def area(self):
        return self.length*self.length
class Circle(Shape):
    def __init__(self,radius):
        super().__init__("圆形")
        self.radius=radius
    def area(self):
        return 3,14*self.radius*self.radius
class Triangle(Shape):
    def __init__(self,base,height):
        super().__init__("三角形")
        self.base=base
        self.height=height
    def area(self):
        return self.base*self.height/2
s=Square(5)
c=Circle(6)
t=Triangle(3,4)
```

```python
class A(str):
    def __init__(self,num):
        self.num=num
    def __int__(self):
        try:
            return int(self.num)
        except ValueError:
            zh={"零":0,"一":1,"二":2,"三":3,"四":4,"五":5,"六":6,"七":7,"八":8,"九":9,"壹":1,"贰":2,"叁":3,"肆":4,"伍":5,"陆":6,"柒":7,"捌":8,"玖":9}
            result=0
            for each in self.num:
                if each in zh:
                    result+=zh[each]
                else:
                    result+=int(each)
                result*=10
            return result//10
```

```python
class Double:
    def __init__(self,start,stop):
        self.value=start-1
        self.stop=stop
    def __iter__(self):
        return self#self本身就是迭代器，所以返回本身
    def __next__(self):
        if self.value==self.stop:
            raise StopIteration#该异常为特殊异常，python不会报错，而是告诉for循环该停止循环了
        self.value+=1
        return self.value*2
d=Double(1,5)
for i in d:
    print(i,end=" ")
```

```python
class C:
    def __init__(self):
        self._x=250
    def __getattr__(self,arg):
        if arg=="x":
            return self._x
        else:
            super().__getattr__(arg)
    def __setattr__(self,arg,value):
        if arg=="x":
            super().__setattr__("_x",value)
        else:
            super().__setattr__(arg,value)
    def __delattr__(self,arg):
        if arg=="x":
            super().__delattr__("_x")
        else:
            super().__delattr__(arg)
c=C()
c.x
```
